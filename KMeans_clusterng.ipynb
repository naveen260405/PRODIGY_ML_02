```python
#importing the necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
#Reading the dataset
df = pd.read_csv('/content/Mall_Customers.csv')
df.head()
```

```python
#Performing Exploratory Data Analysis
#Checking for datatypes
df.dtypes
```

```python
#Checking for null or missing values
df.isnull().sum()
```

```python
#Descriptive statistics
df.describe()
```

```python
#Checking for duplicated values
df.duplicated()
```

```python
#Scatter plot
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
```

```python
#Feature normalisation
scaler = MinMaxScaler()
scaler.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
df[['Annual Income (k$)', 'Spending Score (1-100)']] = scaler.transform(df[['Annual Income (k$)', 'Spending Score (1-100)']])
df
```

```python
#Finding the number of clusters using elbow plot
k_rng = range(1,10)
sse=[]
for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
  sse.append(km.inertia_)
```

```python
sse #Sum of squared error
```

```python
plt.xlabel('k')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse, marker='o', markerfacecolor='red', markersize=9)
```

```python
#Training the K-Means Clustering Model
km = KMeans(n_clusters=5)
km.fit_predict(df[['Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
df['cluster'] = km.fit_predict(df[['Annual Income (k$)', 'Spending Score (1-100)']])
df.head()
```

```python
km.cluster_centers_
```

```python
#Grouping the clusters
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]
plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'], color='brown', label='cluster 1')
plt.scatter(df2['Annual Income (k$)'], df2['Spending Score (1-100)'], color='red', label='cluster 2')
plt.scatter(df3['Annual Income (k$)'], df3['Spending Score (1-100)'], color='blue', label='cluster 3')
plt.scatter(df4['Annual Income (k$)'], df4['Spending Score (1-100)'], color='green', label='cluster 4')
plt.scatter(df5['Annual Income (k$)'], df5['Spending Score (1-100)'], color='purple', label='cluster 5')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1,], color='black', marker='*', label = 'centriod')
plt.legend()
```
